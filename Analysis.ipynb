```python
import pandas as pd

# Load the dataset
df = pd.read_csv('call_center_data.csv')
```

**Conclusion:** This step identifies any missing values in the dataset, helping us understand which columns may need data cleaning.


```python

df.fillna(0, inplace=True)
```

**Conclusion:** Missing values are handled by filling them with 0. This ensures that the dataset is ready for analysis without any null values.


```python

print("\nSummary statistics for key numeric columns:")
print(df[['chatSeconds', 'timeDuration', 'amount', 'astrologersEarnings', 'netAmount', 'userOnCallDuration']].describe())
```

**Conclusion:** Summary statistics provide a quick overview of the central tendency and spread of key numeric columns, offering insights into their distributions.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of call charges
plt.figure(figsize=(10, 6))
sns.histplot(df['amount'], bins=30, kde=True)
plt.title('Distribution of Call Charges')
plt.xlabel('Call Charges')
plt.ylabel('Frequency')
plt.show()
```

**Conclusion:** The distribution plot helps visualize how call charges are spread across different ranges, identifying any potential outliers or common charge amounts.


```python
# Average TalkTime for different call activities
average_talktime = df.groupby('consultationType')['userOnCallDuration'].mean()
print("\nAverage TalkTime for different call activities:")
print(average_talktime)
```

**Conclusion:** This analysis reveals the average TalkTime for each type of consultation, which can be useful for understanding customer engagement for different services.


```python
# Most common source of calls
common_source = df['source'].mode()[0]
print(f"\nThe most common source of calls is: {common_source}")
```

**Conclusion:** Identifying the most common source of calls helps in understanding the primary channels through which customers are reaching out.


```python
# Total earnings and spending
total_earnings = df['astrologersEarnings'].sum()
total_spending = df['amount'].sum()
print(f"\nTotal Earnings: {total_earnings}")
print(f"Total Spending: {total_spending}")
```

**Conclusion:** This step provides a summary of total earnings and spending, giving an overview of the financial aspects of the call center operations.


```python
# Scatter plot for relationship between TalkTime and Charge
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['userOnCallDuration'], y=df['amount'])
plt.title('Relationship between TalkTime and Charge')
plt.xlabel('TalkTime')
plt.ylabel('Charge')
plt.show()
```

**Conclusion:** The scatter plot visualizes the relationship between TalkTime and call charges, helping to identify any correlation between these variables.


```python
# Common reason for call disconnection
common_disconnection_reason = df['callStatus'].mode()[0]
print(f"\nThe most common reason for call disconnection is: {common_disconnection_reason}")
```

**Conclusion:** Identifying the most common reason for call disconnection helps in addressing frequent issues and improving overall call quality.


```python
# Analyze HangUpTime patterns
plt.figure(figsize=(10, 6))
sns.histplot(df['callSeconds'], bins=30, kde=True)
plt.title('HangUpTime Patterns')
plt.xlabel('HangUpTime')
plt.ylabel('Frequency')
plt.show()
```

**Conclusion:** Analyzing HangUpTime patterns can provide insights into typical call durations, helping to optimize call handling procedures.


```python
# Order status distribution for calls
order_status_distribution = df['chatStatus'].value_counts()
print("\nOrder status distribution for calls:")
print(order_status_distribution)
```

**Conclusion:** Understanding the distribution of order statuses helps in identifying common call outcomes and planning for better resource allocation.


```python

# Convert 'refunds[0]' to numeric, coercing errors to NaN
df['refunds[0]'] = pd.to_numeric(df['refunds[0]'], errors='coerce')

# Calculate total refund amount
total_refunds = df['refunds[0]'].sum()

# Calculate refund status distribution
refund_status_distribution = df['refundStatus'].value_counts()

# Print results
print(f"\nTotal Refunds: {total_refunds}")
print(refund_status_distribution)
```

**Conclusion:** This analysis provides insights into the total refunds processed and the distribution of refund statuses, which is essential for financial planning and customer satisfaction.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame containing the 'createdAt' and 'amount' columns
# Sample data or actual data loading should be done before this point

# Convert 'createdAt' to datetime format
df['createdAt'] = pd.to_datetime(df['createdAt'])

# Plotting the trend in call charges over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='createdAt', y='amount', data=df)  # Corrected 'data' parameter
plt.title('Trend in Call Charges Over Time')
plt.xlabel('Time')
plt.ylabel('Call Charges')
plt.show()
```

**Conclusion:** Visualizing the trend in call charges over time helps in identifying any seasonal patterns or trends that could be leveraged for business strategies.


```python
# Relationship between TalkTime and UserSpend
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['userOnCallDuration'], y=df['amount'])
plt.title('Relationship between TalkTime and UserSpend')
plt.xlabel('TalkTime')
plt.ylabel('UserSpend')
plt.show()
```

**Conclusion:** This visualization further explores the relationship between TalkTime and user spending, providing more insights for customer engagement strategies.


```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the 'refundStatus' column
# Sample data or actual data loading should be done before this point

# Calculate value counts for 'refundStatus'
refund_status_counts = df['refundStatus'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(8, 4))
plt.pie(refund_status_counts, labels=refund_status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Refund Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Assuming df is your DataFrame containing the 'consultationType' column
# Sample data or actual data loading should be done before this point

# Calculate value counts for 'consultationType'
consultation_type_counts = df['consultationType'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(8, 5))
consultation_type_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Consultations by Consultation Type')
plt.xlabel('Consultation Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
